```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.cos(x)-x
```

```python
#Metoda bisekce; je nutno zvolit kraje intervalu (a,b) tak, že f(a) má jiné znaménko než f(b)
a=0.1
b=2
n=10 #počet iterací

x=np.linspace(a,b,200)
plt.plot(x,f(x))
plotht=(f(a)-f(b))
plotbt=f(b)

m=(a+b)/2
s=np.sign(f(a))
if np.sign(f(b))!=-s:
    print("Chyba, nesplněn předpoklad: f(a) má stejné znaménko jako f(b)")
for k in range(n):
    if np.sign(f(m))==s:
        a=m
    else:
        b=m
    m=(a+b)/2
    plt.bar(m,plotht,width=(b-a),bottom=plotbt,alpha=0.1,facecolor='black')
    #print(m)
print(m)
```

```python
#Newtonova metoda; je nutné zvolit počáteční bod x[0] a zadat derivaci f
x=[0.1]
def df(x):
    return -np.sin(x)-1
n=10 #počet iterací

for k in range(n):
    x.append(x[k]-f(x[k])/df(x[k]))
    #print(x[k+1])
print(x[k+1])    

col=np.linspace(0,1,n+1)
plt.scatter(x,[0]*(k+2),c=range(k+2))
xx=np.linspace(min(x),max(x),200)
plt.plot(xx,f(xx),color='red')
```

```python
#Metoda sečen; je nutné zvolit nějaké dva počáteční body x[0], x[1]
x=[0.1,0.2]
n=10 #počet iterací

for k in range(1,n):
    #print(x[k])
    x.append((x[k-1]*f(x[k])-x[k]*f(x[k-1]))/(f(x[k])-f(x[k-1])))
    if (x[k]==x[k+1]):
        break
print(x[k+1])

col=np.linspace(0,1,k+2)
plt.scatter(x,[0]*(k+2),c=range(k+2))
xx=np.linspace(min(x),max(x),200)
plt.plot(xx,f(xx),color='red')
```
