```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 1/(x*(x-3))
```

```python
x=[1]
n=10 #počet iterací
plt.figure(figsize=(10,10))

for k in range(n):
    x.append(f(x[k]))
    print(x[k])
    plt.plot([x[k],x[k],x[k+1]],[x[k],x[k+1],x[k+1]],color='lightgrey')
print(x[k])    

b=max(x+[0])
a=min(x+[0])
col=np.linspace(0,1,n+1)
plt.scatter(x[:-1],x[1:],c=range(n))
xx=np.linspace(a,b,200)
plt.xlim([a,b])
plt.ylim([a,b])
plt.plot([a,b],[a,b],color='orange')
plt.plot(xx,f(xx),color='red')
```
