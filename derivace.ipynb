```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)
```

```python
x0=0.2
h=0.1
print("Dopředná diference:",(f(x0+h)-f(x0))/h)
print("Zpětná diference:",(f(x0)-f(x0-h))/h)
print("Symetrická diference:",(f(x0+h)-f(x0-h))/(2*h))
```

```python
def df(x):                        #skutečná derivace
    return np.cos(x)

hh=np.linspace(0.0001,0.1,200)    #vyjadřujeme závislost chyby na délce kroku
err=[(f(x0+h)-f(x0))/h-df(x0) for h in hh]
plt.plot(hh,err)
err=[(f(x0)-f(x0-h))/h-df(x0) for h in hh]
plt.plot(hh,err)
err=[(f(x0+h)-f(x0-h))/(2*h)-df(x0) for h in hh]
plt.plot(hh,err)
```
