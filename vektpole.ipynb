```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

Tento sešit ilustruje možnosti analýzy chování řešení dané diferenciální rovnice pomocí tzv náčrtu vektorového pole, bez toho, abychom znali její řešení.

Uvažujeme tedy obecnou diferenciální rovnici prvního řádu, která je tvaru
$$y'=f(x,y)$$

Tato rovnice nám říká, že každé řešení, jež prochází bodem $(x,y)$ má v tomto bodě směrnici tečny $f(x,y)$. Tuto směrnici tedy můžeme naznačit tak, že do bodu $(x,y)$ umístíme šipku (vektor) ve směru $(1,f(x,y))$. Aby měly všechny šipky stejnou velikost, můžeme je ještě normalizovat. Budeme tedy umisťovat vektor
$$(u,v)=\left({1\over\sqrt{1+f^2(x,y)}},{f(x,y)\over\sqrt{1+f^2(x,y)}}\right)$$

```python
def f(x,y):
    return (x-2)*(np.log(x)-y)
    #return y**2/(1-y)
    #return (y-1)/(x*x-1)
    #return y*(x-y)
```

```python
x,y = np.meshgrid(np.linspace(0.1,4.1,20),np.linspace(-2,2,20))
#x,y = np.meshgrid(np.linspace(-2,2,20),np.linspace(-2,2,20))

u=1/np.sqrt(1+f(x,y)**2)
v=f(x,y)/np.sqrt(1+f(x,y)**2)
plt.quiver(x,y,u,v)
plt.show()
```
